```python
from ultralytics import YOLO

# Load a pre-trained YOLOv8 model (Nano version)
model = YOLO("yolov8n.pt")  # You can also try "yolov8s.pt" for better accuracy

# Train the model using the dataset
model.train(data=f"{dataset_path}/data.yaml", epochs=50, imgsz=640)
```

```python
from ultralytics import YOLO

# Use a slightly larger model
model = YOLO("yolov8s.pt")

# Train with optimizations
model.train(data=f"{dataset_path}/data.yaml",
            epochs=50,
            imgsz=640,
            lr0=0.0005,
            patience=5,
            augment=True)
```
